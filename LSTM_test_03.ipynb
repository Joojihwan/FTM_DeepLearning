```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

df = pd.read_csv('WorldCup.csv', encoding = 'euc-kr')
df
```

```python
df_train = df['00 ~ 01']
```

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
train = df_train

ax = train.plot()
plt.rcParams["figure.figsize"] = (30,10)
plt.legend()
```

```python
def MinMaxScaler(data):
    # 데이터 모든숫자들을 최소 값만큼 뺀다.
    numerator = data - np.min(data, 0)
    # 최대값과 최소 값의 차이(A)를 구한다
    denominator = np.max(data, 0) - np.min(data, 0)
    # 너무 큰 값이 나오지 않도록 나눈다
    return numerator / (denominator + 1e-7)
```

```python
train = MinMaxScaler(train)
train = pd.DataFrame(train, columns = ['00 ~ 01'])
test = train
```

```python
for s in range(1,13):
    train['shift_{}'.format(s)] = train['00 ~ 01'].shift(s)
```

```python
x_train = train.dropna().drop('00 ~ 01',axis=1)
y_train = train.dropna()[['00 ~ 01']]
```

```python
x_train=x_train.values

y_train = y_train.values
print(x_train.shape)
print(y_train.shape)
```

```python
x_train_t = x_train.reshape(x_train.shape[0],12,1)
```

```python
print("최종 데이터")
print(x_train_t.shape)
print(x_train_t)
print(y_train)
```

```python
from keras.layers import LSTM 
from keras.models import Sequential 
from keras.layers import Dense 
import keras.backend as K 
from keras.callbacks import EarlyStopping

K.clear_session()
    
model = Sequential() # Sequeatial Model 
model.add(LSTM(20, input_shape=(12, 1))) # (timestep, feature) 
model.add(Dense(1)) # output = 1 
model.compile(loss='mean_squared_error', optimizer='adam') 
model.summary()
```

```python
early_stop = EarlyStopping(monitor='loss', patience=1, verbose=1)

model.fit(x_train_t, y_train, epochs=200,
          batch_size=30, verbose=1) #, callbacks=[early_stop]
```

```python
data = pd.DataFrame(y_train)

ax = data.plot()
#test.plot(ax=ax)
plt.rcParams["figure.figsize"] = (30,10)
plt.legend()
```
